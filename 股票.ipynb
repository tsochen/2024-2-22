```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 25})
# 資產配置
labels = ['Stocks', 'Bonds', 'Cash']
conservative = [10, 70, 20]
balanced = [50, 40, 10]
growth = [70, 20, 10]
aggressive = [90, 5, 5]

# 繪製圓餅圖
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

axs[0, 0].pie(conservative, labels=labels, autopct='%1.1f%%')
axs[0, 0].set_title('Conservative Portfolio')

axs[0, 1].pie(balanced, labels=labels, autopct='%1.1f%%')
axs[0, 1].set_title('Balanced Portfolio')

axs[1, 0].pie(growth, labels=labels, autopct='%1.1f%%')
axs[1, 0].set_title('Growth Portfolio')

axs[1, 1].pie(aggressive, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Aggressive Portfolio')

plt.tight_layout()
plt.show()
```
